```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
a={'cause of deaths':['Chronic Disease', 'Unintetional Injuries', 'Alzheimers', 'Infuenza and Pneumonia', 'Sepsis','Others'],'Percentile':[62, 5, 4, 2, 1 , 26]}
```

```python
df=pd.DataFrame(data=a)
df
```

```python
df['Percentile'].sum()
```

```python
d=['Chronic Disease', 'Unintetional Injuries', 'Alzheimers', 'Infuenza and Pneumonia', 'Sepsis','Others']
```

```python
plot=df.plot.pie(y='Percentile',labels=d)
```

```python
my_explode=(0,0.1,0.2,0.3,0.4,0)
plt.pie(df['Percentile'],labels=df['cause of deaths'],autopct='%1.1f%%',explode=my_explode)
plt.title('Cause of death')
plt.axis('equal')
plt.show()
```
